```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Drop the columns where all elements are missing or NaN.
properties.dropna(axis=1,how='all',inplace=True)

# Select data from first to 32nd Borough 
properties = properties.iloc[:,:34]
properties = properties.drop(labels='City of London', axis=1) 

print(properties.head())

# Transpose table
properties = properties.transpose()

# Rename columns
properties = properties.reset_index()
properties.iloc[0,0] = 'lbrh'
properties.iloc[0,1] = 'id'
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Assign values of first row and rename index columns
properties.columns = properties.iloc[0]

# Drop duplicate row
properties = properties.drop(labels=0, axis=0)
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt dataframe
properties = pd.melt(properties, id_vars=['lbrh', 'id'], var_name='date', value_name='avg_prc')
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Set price column to float
pd.to_numeric(properties['avg_prc'], downcast='float')

# Extract year column from date
properties['year'] = properties['date'].dt.year

# Check data types
print(properties.dtypes)

# Check dataframe
properties.head()
```

**2.5. Cleaning the data (part 3)**

```python
# Checking the 32 unique London Boroughs are all included
brhs_32 = list(set(properties['lbrh']))

print("Number of unique London Boroughs in data: ", len(set(properties['lbrh'])))
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(dfg, y1, y2):
    """This function calculates the ratio of house prices in Borough dfg, 
    comparing the price of a house in y2 to the price in y1."""
    spec_brh_y1 = properties[(properties['lbrh']==dfg)&(properties['year']==y1)]
    spec_brh_y2 = properties[(properties['lbrh']==dfg)&(properties['year']==y2)]
    
    y1_avg_prc = spec_brh_y1['avg_prc'].agg(np.mean)
    y2_avg_prc = spec_brh_y2['avg_prc'].agg(np.mean)
    price_ratio = [dfg, y1_avg_prc/y2_avg_prc, y1, y1_avg_prc, y2, y2_avg_prc]
    return price_ratio

print(create_price_ratio('Westminster', 1998, 2018))
```

```python
# Create dataframe for two decade comparisons, where data for all months from Jan-Dec are available i.e.1995-2020

a_val = pd.DataFrame(create_price_ratio(i, j, j+20) for i in brhs_32 for j in range(1995,2001,1))
a_val.columns = ['lbrh', 'prc_r', 'y1', 'y1_avg_prc', 'y2', 'y2_avg_prc']
b_val = a_val.sort_values(by=['y1','prc_r'], ascending=[True, True])

# Print df for checking
print(b_val[b_val['y2']==2020])
```

```python
#Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Plot two decade price growth over time
sns.lineplot(data=b_val, y='prc_r', x='y2', hue='lbrh', alpha=0.5)

plt.xlabel('Two Decades from Year')
plt.ylabel('Price Ratio')
plt.legend(title = 'Boroughs of London', loc='upper right', bbox_to_anchor=(1.6, 1.6))
plt.title('Price Growth Over Two Decades')
plt.show()
```

```python
# Filter for three highest average house price growth boroughs for each year over span of two decades, i.e. lowest price ratio

three_high_grow_twodec = pd.DataFrame()
for j in range (1995,2001,1):
    tb1 = pd.DataFrame(b_val[(b_val['y1']==j)].nsmallest(3, 'prc_r'))
    three_high_grow_twodec = three_high_grow_twodec.append(tb1)

print(set(three_high_grow_twodec['lbrh']))
print("\n")

print(three_high_grow_twodec[three_high_grow_twodec['y2']>=2015])
```

```python
sns.scatterplot(data=three_high_grow_twodec, x='y2', y='prc_r', hue='lbrh', alpha=1)

plt.xlabel('Two Decades from Year')
plt.ylabel('Average House Price Ratio')
plt.legend(title = 'Boroughs of London', loc='upper right', bbox_to_anchor=(1.5, 1))
plt.title('Top 3 Highest Average House Price Growth Boroughs')

plt.show()
```

```python
# Filter for three lowest average house price growth boroughs for each year over span of two decades, i.e. highest price ratio
    
three_low_grow_twodec = pd.DataFrame()
for j in range (1995,2001,1):
    tb1 = pd.DataFrame(b_val[(b_val['y1']==j)].nlargest(3, 'prc_r'))
    three_low_grow_twodec = three_low_grow_twodec.append(tb1)

print(set(three_low_grow_twodec['lbrh']))
print("\n")

print(three_low_grow_twodec[three_low_grow_twodec['y2']==2020])
```

```python
# Create dataframe for year on year comparisons, where data for all months from Jan-Dec are available i.e.1995-2020

c_val = pd.DataFrame(create_price_ratio(i, j-1, j) for i in brhs_32 for j in range(1996,2021,1))
c_val.columns = ['lbrh', 'prc_r', 'y1', 'y1_avg_prc', 'y2', 'y2_avg_prc']
d_val = c_val.sort_values(by=['y2','prc_r'], ascending=[True, True])
```

```python
# Print df for checking
print(d_val[d_val['y2']==2020])
```

```python
# Filter for three highest average house price growth boroughs of each year from 1995 to 2020, i.e. lowest price ratio
    
three_high_grow_yoy = pd.DataFrame()
for j in range (1995,2021,1):
    tb1 = pd.DataFrame(d_val[(d_val['y2']==j)].nsmallest(3,'prc_r'))
    three_high_grow_yoy = three_high_grow_yoy.append(tb1)

print(three_high_grow_yoy[three_high_grow_yoy['y2']>=2015])
```

```python
# Filter for three most desirable boroughs from 1995 to 2020, i.e. highest average house price of each year     

three_high_price = pd.DataFrame()
for j in range (1995,2021,1):
    tb1 = pd.DataFrame(d_val[(d_val['y2']==j)].nlargest(3,'y2_avg_prc'))
    three_high_price = three_high_price.append(tb1)

print(three_high_price[three_high_price['y2']>=2015])
```

```python
# Filter for three least desirable boroughs, i.e. lowest average house price of each year 

three_low_price = pd.DataFrame()
for j in range (1995,2021,1):
    tb1 = pd.DataFrame(d_val[(d_val['y2']==j)].nsmallest(3,'y2_avg_prc'))
    three_low_price = three_low_price.append(tb1)

print(three_low_price[three_low_price['y2']>=2015])
```

```python
# Is there a correlation between price ratio and average house price?

sns.scatterplot(data=d_val, x='prc_r', y='y2_avg_prc', hue='lbrh', alpha=0.5)

plt.ylabel('Average House Price')
plt.xlabel('Price Ratio')
plt.legend(title = 'Boroughs of London', loc='upper right', bbox_to_anchor=(1.6, 1.6))
plt.title('Price Growth vs Average House Price')
plt.yscale('log')
plt.show()
```
